```python
import openfoodfacts

# User-Agent is mandatory
api = openfoodfacts.API(user_agent="MyAwesomeApp/1.0")
```

```python
results = api.product.text_search("pizza")
```

```python
# Check if results exist
if results["count"] > 0 and "products" in results:
    product = results["products"][1]
    name = product.get("product_name", "Unknown")
    protein = product.get("nutriments", {}).get("proteins_100g")

    print(f"Product: {name}")
    print(f"Protein per 100g: {protein} g")
else:
    print("No pizza found.")
```

```python
import pandas as pd
import difflib

# Load and normalize CSV
_df = pd.read_csv("../data/food_db.csv")
_df.columns = [c.strip().lower().replace(" ", "_") for c in _df.columns]

def food_analysis(food: str, g: int = None, portions: int = None):
    food_list = _df["food"].tolist()
    most_similar = difflib.get_close_matches(food, food_list, n=1, cutoff=0.5)
    if not most_similar:
        return {"error": "No similar food found"}

    row = _df[_df["food"] == most_similar[0]].iloc[0]
    base_grams = row.get("grams")
    try:
        base_grams = float(str(base_grams).replace(",", ""))
    except:
        return {"error": "Invalid gram value in data"}

    if g is not None:
        scale = g / base_grams
    elif portions is not None:
        scale = portions
    else:
        scale = 1.0

    result = {
        "name": row["food"],
        "measure": row["measure"],
        "scaled_grams": round(base_grams * scale, 2),
    }

    for col in _df.columns:
        if col in ["food", "measure", "grams", "category"]:
            continue
        val = row[col]
        try:
            val = float(str(val).replace(",", ""))
            result[col] = round(val * scale, 2)
        except:
            result[col] = val  # keep 't' or non-numeric values as-is

    return result

res = food_analysis("teryaki", g=100)
res
```
